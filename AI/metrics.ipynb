```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# Recall 데이터만 사용
labels = ['RT-DETR', 'FINE_TUNED']
scores = [44.8, 69.9]

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots(figsize=(8, 6))

# 막대 (각각 따로 생성해서 legend 유지)
bars1 = ax.bar(x[0], scores[0], width, label='RT-DETR', color='#FFA07A')
bars2 = ax.bar(x[1], scores[1], width, label='FINE_TUNED', color='#87CEEB')

# 제목 & 축
ax.set_title('Recall(재현율) 성능 비교', fontsize=16, fontweight='bold')
ax.set_ylabel('점수 (%)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 100)

# 그리드
ax.grid(axis='y', linestyle='--', alpha=0.6)

# 수치 표시
for bar in list(bars1) + list(bars2):
    height = bar.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5),
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=12, fontweight='bold')

# 🔥 Recall 증가 화살표
ax.annotate('놓치던 객체를 더 많이 검출',
            xy=(x[1], scores[1]),
            xytext=(x[1] + 0.3, scores[1] + 15),
            arrowprops=dict(arrowstyle='->', color='red', lw=2),
            fontsize=13, fontweight='bold', color='red')

# ✅ legend 복구
ax.legend()

# 여백
ax.margins(x=0.2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# Recall 데이터만 사용
labels = ['RT-DETR', 'FINE_TUNED']
scores = [97.4, 94.5]

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots(figsize=(8, 6))

# 막대 (각각 따로 생성해서 legend 유지)
bars1 = ax.bar(x[0], scores[0], width, label='RT-DETR', color='#FFA07A')
bars2 = ax.bar(x[1], scores[1], width, label='FINE_TUNED', color='#87CEEB')

# 제목 & 축
ax.set_title('Precision(정밀도) 성능 비교', fontsize=16, fontweight='bold')
ax.set_ylabel('점수 (%)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(60, 100)

# 그리드
ax.grid(axis='y', linestyle='--', alpha=0.6)

# 수치 표시
for bar in list(bars1) + list(bars2):
    height = bar.get_height()
    ax.annotate(f'{height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5),
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=12, fontweight='bold')

# 🔥 Recall 증가 화살표
ax.annotate('오탐지가 소폭 증가',
            xy=(x[1] + 0.1, scores[1]),
            xytext=(x[1] + 0.3, scores[1] + 15),
            arrowprops=dict(arrowstyle='->', color='red', lw=2),
            fontsize=13, fontweight='bold', color='red')

# ✅ legend 복구
ax.legend()

# 여백
ax.margins(x=0.2)

plt.tight_layout()
plt.show()
```
